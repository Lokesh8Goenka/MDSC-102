```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
name = ["Class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
df = pd.read_csv('./wine/wine.data' ,names = name)
df
```

```python
sns.countplot(x = "Alcohol" , data = df)
```

```python
sns.scatterplot(x = "Alcohol", y = "Malic acid", hue= 'Class', data = df)
```

```python
plt.figure(figsize=(20,20))
for index, column in enumerate(df):
    if index > 1 : 
        plt.subplot(4,3,index - 1)
        sns.scatterplot(df, x='Alcohol', y=column, hue='Class')

plt.show()
```
