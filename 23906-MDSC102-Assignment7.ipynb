```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
name = ["Class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
df = pd.read_csv('./wine/wine.data', names = name)
df
```

```python
df2 = df.drop(['Class'], axis=1)
df2
```

```python
fig,axes=plt.subplots(1,2,figsize=(20,70))
j = 1

features  = ["Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
for i in range(len(features)):

        plt.title("Plot for {}".format(features[i]))
        plt.subplot(13,3,j)
        sns.boxplot(data= df2, y = features[i])
        j += 1

        plt.title("Plot for {}".format(features[i]))
        plt.subplot(13,3,j)
        sns.histplot(data = df2, x = features[i], kde = True, hue = df2[features[i]].skew())
        j += 1

        plt.title("Plot for {}".format(features[i]))
        plt.subplot(13,3,j)
        sp.stats.probplot(df2[features[i]], plot = plt)
        j += 1

        
plt.show() 
```

* Alcohol = Normal
* Malic acid =  Right skewed, as the skew value is 1.03 and the right viscous is larger than the left one.
* Ash =  Normal
* Alcalinity of ash = Normal
* Magnesium = Right skewed, as the skew value is 1.09 and the right viscous is larger than the left one.
* Total phenols =  Right skewed as the right viscous is larger than the left one.
* Flavanoids =  Right skewed as the right viscous is larger than the left one.
* Nonflavanoid phenols = Right skewed as the right viscous is larger than the left one.
* Proanthocyanins =  Right skewed as the right viscous is larger than the left one.
* Color intensity =  Right skewed as the right viscous is larger than the left one.
* Hue =  Right skewed as the right viscous is larger than the left one.
* OD280/OD315 of diluted wines =  Left skewed as the skew value is -0.30 and the box plot is also left skewed.
* Proline = Right skewed as the right viscous is larger than the left one.
