```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
name = ["Class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
df = pd.read_csv('./wine/wine.data' ,names = name)
df
```

```python
df.info()
df.shape
```

```python

df.describe()
```

```python
plt.figure(figsize=(15,15))

plt.subplot(3,2,1)
sns.boxplot(data= df, y = "Magnesium")
plt.title("Magnesium level")

plt.subplot(3,2,2)
sns.histplot(data = df, x = "Magnesium", kde = True)
plt.title("Magnesium level")

plt.subplot(3,2,3)
sns.boxplot(data= df, y = "OD280/OD315 of diluted wines")
plt.title("Level of OD280/OD315 of diluted wines")

plt.subplot(3,2,4)
sns.histplot(data = df, x ="OD280/OD315 of diluted wines", kde = True)
plt.title("diluted wines")


plt.subplot(3,2,5)
sns.boxplot(data= df, y = "Hue")
plt.title("Level of  Hue")


plt.subplot(3,2,6)
sns.histplot(data = df, x ="Hue", kde = True)
plt.title("Level of Hue")

plt.savefig("./")
plt.show()
```

```python
Magnesium1 = np.log(df["Magnesium"]) # log of the original magnesium data
Magnesium1
```

```python
Magnesium1.skew()
```

```python
Magnesium2 = np.sqrt(df["Magnesium"]) # sqrt of the original magnesium value
#Magnesium2
Magnesium2.skew()
```

```python
Magnesium3 = 1/(df["Magnesium"])
Magnesium3.skew()

mag,lamb = sp.stats.boxcox(df["Magnesium"])

mag_boxcox = ((df["Magnesium"]**lamb) - 1) / lamb
print(mag_boxcox)
mag_boxcox.skew()

```

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))

s=sp.stats.probplot(df["Magnesium"], plot = axes[0])
axes[0].set_title = ("Original Magnesium Values")

s=sp.stats.probplot(mag_boxcox, plot = axes[1])
axes[1].set_title = ("Transformed Magnesium Values")

```

```python
plt.figure(figsize = (15,5)) 
plt.subplot(1,2,1)
sns.histplot(data = mag_boxcox, x = mag_boxcox, kde = True)
plt.title("Original Magnesium Values")

plt.subplot(1,2,2)
sns.histplot(data = df["Magnesium"], x = df["Magnesium"], kde =True)
plt.title("Transformed Magnesium Values")
```
