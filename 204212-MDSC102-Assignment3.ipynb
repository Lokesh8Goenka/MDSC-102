## Plotting different distributions


```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Binomial distribution

 $ B(n,p) = (C(n x)) p(1-p)^{n-x} $ 

```python
def B(n,p):
    f = lambda n,p,r: (np.math.factorial(n) / (np.math.factorial(r) * np.math.factorial(n - r)) * (p**r) * ( (1-p)**(n-r) )  )
    r = np.arange(0, n+1, dtype=int) # x values
    y = [ f(n,p,i) for i in r]
    plt.plot(r, y)

    plt.title("Binomial distribution")
    plt.xlabel("x values")
    plt.ylabel("F(x)")
    plt.legend(["n = 30, p = 0.2"])
    plt.show()    

B(30, 0.2)
```

## Poisson

$ p(λ) = e^-λ (λ^x)/x! $

```python
def P(l):
    f = lambda l,x: np.exp(-l) * ((l**x)/np.math.factorial(x))
    r = np.arange(0, 30, dtype=int) # x values
    y = [ f(l,i) for i in r]
    plt.plot(r, y)

    l = 3
    y = [ f(l,i) for i in r]
    plt.plot(r, y)

    l = 8
    y = [ f(l,i) for i in r]
    plt.plot(r, y)

    plt.title("Poisson distribution")
    plt.xlabel("x values")
    plt.ylabel("F(x)")
    plt.legend(["λ = 1", "λ = 3", "λ = 8"])
    plt.show()    

P(1)
```

## Exponential


$f(x) = λ e^{-λx}$

```python
def Expo(l):
    f = lambda l,x: l * np.exp(-l*x)
    r = np.arange(0, 30, dtype=int) # x values
    y = [ f(l,i) for i in r]        # y values
    plt.plot(r, y)

    l = 0.01
    y = [ f(l,i) for i in r]        # y values
    plt.plot(r, y)

    l = 5
    y = [ f(l,i) for i in r]        # y values
    plt.plot(r, y)

    plt.title("Exponential distribution")
    plt.xlabel("x values")
    plt.ylabel("F(x)")
    plt.legend(["λ = 2", "λ = 0.01", "λ = 5"])
    plt.show()    

Expo(2)
```

## Normal Distribution

```python
def normal(s,mu):
    f = lambda s,mu,x: (1/(np.sqrt(2*np.pi*s))) * np.exp(-0.5*( (x-mu)**2/s) )
    r = np.arange(-50, 50, dtype=int) # x values
    y = [ f(s,mu,i) for i in r]        # y values
    plt.plot(r, y)

    s = 2
    mu = 10
    y = [ f(s,mu,i) for i in r]        # y values
    plt.plot(r, y)

    s = 5
    mu = -15
    y = [ f(s,mu,i) for i in r]        # y values
    plt.plot(r, y)
    
    plt.title("Normal distribution")
    plt.xlabel("x values")
    plt.ylabel("F(x)")
    plt.legend(["σ = 1, μ  = 0", "σ = 2, μ  = 10", "σ = 5, μ  = -15"])
    plt.show()    

normal(1,0)
```

## Gamma Distribution

$f(x; alpha,λ) = (λ^α x^{λ-1})/gama function(α) * e^{-λx} $ 

```python
def Gamma(a,l):
    f = lambda a,l,x: ((l**a) * (x**(l-1)) * np.exp(-l*x)) / math.gamma(a-1)
    r = np.linspace(0,10,100) # x values
    y = [ f(a,l,i) for i in r]        # y values
    plt.plot(r, y)

    a = 4
    l = 6
    y = [ f(a,l,i) for i in r]        # y values
    plt.plot(r, y) 

    a = 3
    l = 7
    y = [ f(a,l,i) for i in r]        # y values
    plt.plot(r, y) 

    plt.title("Gamma distribution")
    plt.xlabel("x values")
    plt.ylabel("F(x)")
    plt.legend(["α = 5,λ = 3", "α = 4,λ = 6", "α = 3,λ = 7"])
    plt.show()    

Gamma(5,3)
```
