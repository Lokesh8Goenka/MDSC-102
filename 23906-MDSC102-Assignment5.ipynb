## BoxPlot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
name = ["Class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
df = pd.read_csv('./wine/wine.data' ,names = name)
df
```

```python
sns.boxplot(data = df, x = "Malic acid")
```

```python
sns.histplot(data = df, x = "Magnesium")
```

```python
plt.figure(figsize=(20,20))
j = 1
for i in name[1:]:
        plt.subplot(4,4,j)
        sns.boxplot(df, x='Class', y=i, hue='Class')
        j += 1

plt.show()        
```

```python
plt.figure(figsize=(40,80))
j = 1
k = 1
x = range(1,4)
for k in x:
        for i in name[1:]:

                plt.subplot(13,3,j)
                plt.title("{} of class {}" .format(i,k))
                sns.histplot(df[df["Class"] == k][i], kde = True)
                j += 1

plt.show()        
```
